```python
import pandas as pd
import numpy as np

df = pd.read_excel("/content/data.xlsx")

alternatives = df.iloc[:, 0]
X = df.iloc[:, 1:].values.astype(float)

weights = np.array([0.25, 0.20, 0.20, 0.15, 0.20])
impact  = np.array([1, -1, 1, 1, -1])
norm_X = X / np.sqrt((X ** 2).sum(axis=0))

weighted_X = norm_X * weights

ideal_best = np.where(impact == 1,
                      weighted_X.max(axis=0),
                      weighted_X.min(axis=0))

ideal_worst = np.where(impact == 1,
                       weighted_X.min(axis=0),
                       weighted_X.max(axis=0))

dist_best = np.sqrt(((weighted_X - ideal_best) ** 2).sum(axis=1))
dist_worst = np.sqrt(((weighted_X - ideal_worst) ** 2).sum(axis=1))

df["TOPSIS Score"] = dist_worst / (dist_best + dist_worst)
df["Rank"] = df["TOPSIS Score"].rank(ascending=False).astype(int)

print(df)
```
